```python

import numpy as np
import pandas as pd

df = pd.read_csv('/content/ethnic diversity.csv')
print('\nLoading the dataset :\n',df.head())
```

```python
# data cleaning

print('\nData information :\n',df.info())

print('\nShape of the data :\n',df.shape)

print('\nChecking null values :\n',df.isnull().sum())

print('\nChecking duplicates :\n',df.duplicated().sum())

print('\nDescriptive  statistics :\n',df.describe())
```

```python
# adding a new column
df['City'] = 'USA'
print(df.head())

# Basic filtering
df_high_age = df[df['age'] > 50]
print('\nHigh_age :\n',df_high_age.tail())

#using and operator
df_age = df[(df['age'] >= 33) & (df['age'] < 50)]
print('\nAge between 33 and 50 :\n',df_age.head())

# using or operator
df_age_or = df[(df['age'] > 70) | (df['age'] >= 35)]
print('\nAge : \n',df_age_or.tail())

# # Widely used methods in this format

df['Zip'] = df['Zip'].astype('int64')
print('Changing data type :\n',df.dtypes)

df_sort = df['age'].sort_values(ascending=False)
print('\nSorting on descending order :\n',df_sort.head())

df_count = df['EmpID'].value_counts().sum()
print('\nTotal employee :\n',df_count)

df_group = df.groupby('Employee_Name')['Salaries'].mean()
print('\nGrouping 2 columns :\n',df_group.head())

df_emp = ['Brown, Mia','Favis, Donald','Saada, Adell']
filtered_data = df[df['Employee_Name'].isin(df_emp)]
print('\nFiltered data :\n',filtered_data)

df['Position'] = df['Position'].str.upper()
print('\nPostion column in upper case :\n',df.head())

def Age(f):
  if 33 <= f < 40:
    return 'Young'
  else:
    return 'Old'
df['Category'] = df['age'].apply(Age)
print('\nCategory column :\n',df.head())

print('\n1st indexing :\n',df.loc[1])
print('\nFrom 2nd index to 4th index :\n',df.loc[2:4])
print('\nFrom 2nd row to 5th row with specific columns :\n',df.loc[2:5,['EmpID','Salaries','City']])

print('\n2nd index using iloc :\n',df.iloc[2])
print('\nFrom 2nd index to 4th index using iloc :\n',df.iloc[2:5])
print('\nFrom 2nd row to 5th row with specific columns using iloc :\n',df.iloc[2:6,1:4])
```

Numerical Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/ethnic diversity.csv')
print(df.head())

# Exploratory Data Analysis

# measure of central tendency
mean_of_age = df.age.mean()
print('\nMean of age :\n',mean_of_age)

median_of_age = df.age.median()
print('\nMedian of age :\n',median_of_age)

mode_of_age = df.age.mode()
print('\nMode of age :\n',mode_of_age)

# measure of dispersion
# variance
variance_of_salary = df.Salaries.var()
print('\nVariance of salary :\n',variance_of_salary)

#standard deviation
std_of_salary = df.Salaries.std()
print('\nStandard deviation of salary :\n',std_of_salary)

# range
range_of_salary = df.Salaries.max() - df.Salaries.min()
print('\nRange of salary :\n',range_of_salary)

# third moment business decision
#skewness
print('\nSkewness of age :\n',df.age.skew())
print('\nSkewness of salary :\n',df.Salaries.skew())

# 4th moment of bussiness decision
# kurtosis
print('\nKurtosis of age :\n',df.age.kurt())
print('\nKurtosis of salary :\n',df.Salaries.kurt())
```

Normalization




```python
# Min-Max Normalization
df_norm = df.copy()
for n in ['age','Salaries']:
  min_val = df[n].min()
  max_val = df[n].max()
  df_norm[n] = (df[n] - min_val) / (max_val - min_val)
print('Min-Max Normalization \n:',df_norm)
```

Z-score Standardization

```python
df_z = df.copy()
for Z in ['age','Salaries']:
  mean_val = df[Z].mean()
  std_val = df[Z].std()
  df_z[Z] = (df[Z] - mean_val) / std_val
print('Z-score Standardization :\n',df_z.head())
```

imputation

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer

#mean imputation
mean_imputer = SimpleImputer(strategy='mean')
df[['Zip','age','Salaries']] = mean_imputer.fit_transform(df[['Zip','age','Salaries']])
print("\nData frame after mean imputation :\n",df)

#mode imputation
mode_imputer = SimpleImputer(strategy='most_frequent')
df[['Zip','age','Salaries']] = mode_imputer.fit_transform(df[['Zip','age','Salaries']])
print("\nData frame after mode imputation :\n",df)
```

Visualization for imputation

```python
fig,axes = plt.subplots(1,2,figsize=(10,6))

#before imputation
sns.heatmap(pd.DataFrame(df).isnull(),cbar=False,ax=axes[0],cmap='Reds')
axes[0].set_title('Before Imputation')

# after imputation
sns.heatmap(df.isnull(),cbar=False,cmap='Reds',ax=axes[1])
axes[1].set_title('After Imputation')
plt.tight_layout()
plt.show()
```


Binarization

```python
from sklearn.preprocessing import Binarizer

binarizer_zip = Binarizer(threshold=2703)
binarizer_age = Binarizer(threshold=33)
binarizer_salary = Binarizer(threshold=27603)

df['zip_code'] = binarizer_zip.fit_transform(df[['Zip']])
df['Age'] = binarizer_age.fit_transform(df[['age']])
df['salary'] = binarizer_salary.fit_transform(df[['Salaries']])
```

Discretization

```python
df['Age_Group'] = pd.cut(df['age'],bins=[18,32,45,60],labels=['18-30','31-45','60+'])

df['Salary_Bins'] = pd.cut(df['Salaries'],bins=[500,1000,15000,50000],labels=['Low','Medium','High'])

df['Age_Decades'] = pd.cut(df['age'],bins=[20, 30, 40, 60],labels=['20s','30s','50s'])
```

```python
fig,axes = plt.subplots(3,2,figsize=(25,25))

#Age_Group
sns.histplot(df['age'],bins=4,kde=True,ax=axes[0,0],color='skyblue')
axes[0,0].axvline(200,color='red',linestyle='--',linewidth=1.5)
axes[0,0].set_title('Age',fontsize=25)
sns.histplot(x=df['Age_Group'],ax=axes[0,1],palette='viridis')
axes[0,1].set_title('Age Group after discretization',fontsize=25)

#Salary_Bins
sns.histplot(df['Salaries'],kde=True,bins=4,ax=axes[1,0],color='skyblue')
axes[1,0].axvline(200,color='red',linestyle='--',linewidth=1.5)
axes[1,0].set_title('Salary',fontsize=25)
sns.histplot(x=df['Salary_Bins'],ax=axes[1,1],palette='magma')
axes[1,1].set_title('Salary Bins after discretization',fontsize=25)

#Age_Decades
sns.histplot(df['age'],bins=4,kde=True,ax=axes[2,0],color='skyblue')
axes[2,0].axvline(200,color='red',linestyle='--',linewidth=1.5)
axes[2,0].set_title('Age Decades',fontsize=25)
sns.histplot(x=df['Age_Decades'],ax=axes[2,1],palette='viridis')
axes[2,1].set_title('Age Decades after discretization',fontsize=25)

plt.tight_layout()
plt.show()
```
